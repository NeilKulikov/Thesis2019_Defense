```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.interpolate as inter
```

```python
xs = [0.3, 0.65, 0.8, 1.0]
dsi = [5.2, 8.7, 9.1, 9.4]
dsp = [2.4, 4.4, 4.7, 5.0]
eti = [16., 15., 15., 13.]
etp = [23., 38., 37., 36.]
```

```python
dsis = inter.CubicSpline(xs, dsi)
dsps = inter.CubicSpline(xs, dsp)
etis = inter.CubicSpline(xs, eti)
etps = inter.CubicSpline(xs, etp)
```

```python
xsn = np.arange(0.3, 1., 1e-3)
dsin = dsis(xsn)
dspn = dsps(xsn)
etin = etis(xsn)
etpn = etps(xsn)
```

```python
plt.figure(figsize=(10,5), dpi = 120)
fig, ax1 = plt.subplots()
ax1.plot(xsn, dsin, 'r-', label = r'$d(x)$, impure')
ax1.plot(xsn, dspn, 'r--', label = r'$d(x)$, pure')
ax1.set_ylim((0., 10.))
ax1.tick_params('y', colors='r')
ax1.grid()
ax1.set_xlabel(r'x')
ax1.set_ylabel(r'$d$, nm')
ax2 = ax1.twinx()
ax2.plot(xsn, etin, 'b-', label = r'$\varepsilon_{th}(x)$, impure')
ax2.plot(xsn, etpn, 'b--', label = r'$\varepsilon_{th}(x)$, pure')
ax2.set_ylim((0., 40.))
ax2.tick_params('y', colors='b')
ax2.set_ylabel(r'$\varepsilon_{th}$, meV')
fig.legend()
plt.savefig('')
```
