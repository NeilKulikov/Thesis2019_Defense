```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import csv
```

```python
k = []
c = [[], []]
v = [[], []]
with open('./18u_impure_80K.dat', 'r') as f:
    reader = csv.reader(f, delimiter = '\t')
    for ki, v2, v1, c1, c2 in reader:
        try:
            k.append(float(ki))
            c[0].append(float(c1))
            c[1].append(float(c2))
            v[0].append(float(v1))
            v[1].append(float(v2))
        except:
            continue
```

```python
ks = np.array(k)
cs = [np.array(ci) for ci in c]
vs = [np.array(vi) for vi in v]
```

```python
plt.plot(ks, cs[0])
```

```python
import scipy.interpolate as inter
```

```python
def get_more_points(ys, xs = ks, nx = 1000):
    spl = inter.interp1d(xs, ys, kind = 'cubic')
    nxs = np.linspace(0., max(xs), nx)
    nys = np.array([spl(nx) for nx in nxs])
    return (nxs, nys)
```

```python
def find_nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return idx
```

```python
plt.figure(figsize = (5,5), dpi = 120)
plt.grid()
nx, ny = get_more_points(cs[0])
plt.plot(nx, ny, 'b-', label = r'$C_1$ branch')
plt.plot(- nx, ny, 'b-')
ex = find_nearest(nx, 0.005)
plt.arrow(0.005, ny[ex], 0.194, ny[find_nearest(nx, 0.198)])
plt.plot(0.005, ny[ex], 'bx', label = r'Initial C state')
ex = find_nearest(nx, 0.198)
plt.plot(0.198, ny[ex], 'bo', label = r'Final C state')
nx, ny = get_more_points(cs[1])
plt.plot(nx, ny, 'g-', label = r'$C_2$ branch')
plt.plot(- nx, ny, 'g-')
nx, ny = get_more_points(vs[0])
plt.plot(nx, ny, 'r-', label = r'$V_1$ branch')
plt.plot(- nx, ny, 'r-')
hx = find_nearest(nx, 0.208)
plt.arrow(-0.203, ny[hx], 0.208, -ny[hx])
plt.plot(-0.208, ny[hx], 'rx', label = 'Initial V state')
nx, ny = get_more_points(vs[1])
plt.plot(nx, ny, 'y-', label = r'$V_2$ branch')
plt.plot(- nx, ny, 'y-')
plt.xlabel(r'$k,nm^{-1}$')
plt.ylabel(r'$\varepsilon,meV$')
plt.xlim(-0.5, 0.5)
plt.ylim(-150., 300.)
plt.legend()
plt.savefig('/Users/neilkulikov/Thesis2019/Thesis2019/images/18u_impure_40K.pdf', bbox_inches = 'tight')
```
